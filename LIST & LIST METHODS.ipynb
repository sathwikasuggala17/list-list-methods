* lists are mutable
* lists allows duplicates
* lists are heterogenous-it can take any value inside it.
* lists are sequential-which have an index value.
* lists are mutable-it can change the elements within the list of same memory location.
* lists are dynamic in nature
* when calling a method results in updation of list, these updation are reflected in same memory location and are applied on same variable, unlike string, we don't have reassign to make them reflect original var
* we can create a list in two ways 1)LIST-the class 2)square braces[]

### list slicing()
* for positive slicing start is zero

* stepsize is plus one or 1.

```python
 #positive indexing - left to right (starts with 0)
y = 'string'
y[0], y[::], y[::-1] #reverse the strings-y[::-1]
```

```python
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'] #slicing the list
print(L[2:7])
```

by using positive indexing:-


```python
#Slice with Negative Indexing
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[-7:-2])
```

by using negative indexing:-


```python
# Slice with Positive & Negative Indices
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[2:-5])
```

### range()
* The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.
* like in strings, we can have a negative step in range
* default value of start in range is zero
* default value of stepsize is +1.
* * range() populates a number within a given range.
* syntax : range(start, stop, step) # end means till end.
> parameters: 
* start	: Optional. An integer number specifying at which position to start. Default is 0
* stop : Required. An integer number specifying at which position to stop (not included end).
* step : Optional. An integer number specifying the incrementation. Default is 1

```python
r = range (1,11)
print(r, type(r), len(r), hex(id(r)), sep = '\n')
```

### copy()
* You cannot copy a list simply by typing list2 = list1, because: list2 will only be a reference to list1, and changes made in list1 will automatically also be made in list2.
* Sometimes, there is a need to reuse any object, hence copy methods are always of great utility.
* There are ways to make a copy, one way is to use the built-in List method copy().
* Another way to make a copy is to use the built-in method list().
* return: return a shallow copy of the list (modification in the new list won't be reflected in the original list.
* syntax : list.copy()
> no parameters

```python
help(list.copy)
```

### 1)shallow copy()
* A shallow copy creates a new object which stores the reference of the original elements.
* So, a shallow copy doesn't create a copy of nested objects, instead it just copies the reference of nested objects. This means, a copy process does not recurse or create copies of nested objects itself.
* syntax : copy.copy(x)
* return as a shallow copy of x.

```python
# shallow copy
x =[1,2,3]
print(x, hex(id(x)))
y=x
print(y, hex(id(y))) # both x and y same memory location
x.append(4)
print(y)
print(hex(id(x[0])), hex(id(y[0]))) #same memory location
```

```python
z = x.copy()
x.append(5)
print(z, hex(id(z)))
```

```python
hex(id(z[0]))
```

```python
a = [1, [4,5,6]]
print(a,a[1], hex(id(a)), hex(id(a[1])))
b = a.copy()
print(b,b[1], hex(id(b)), hex(id(b[1])))
```

```python
a[1].append(7)
a
```

```python
b
```

### 2)deep copy()
* A deep copy creates a new object and recursively adds the copies of nested objects present in the original elements.
* In the case of deep copy, a copy of the object is copied into another object.
* It means that any changes made to a copy of the object do not reflect in the original object.
* It is used to copy nested list
* syntax : copy.deepcopy(x)
* return as a deepcopy of x.

```python
# to handle nested list deep copy is used
import copy
from copy import deepcopy
```

```python
a = [1, [4,5,6]] #diffirent memory location to b
b = deepcopy(a) # different memory location to a
print(a, id(a), id(a[1]))
print(b, id(b), id(b[1]))
```

```python
a[1].append(7)
a
```

```python
b
```

### sort()

* syntax : list.sort(reverse=True|False, key=myFunc)

2) key : Optional. A function to specify the sorting criteria(s)

```python
# to convert a char/int into descending order reverse = true 
cars = ['rolls roys', 'ford', 'thar', 'skoda'] 
cars.sort(reverse=True) 
cars 
```

```python
# reverse=false is for ascending order
n = [10,-80,960,100.1001]
n.sort(key=None, reverse = False)
n # output is ascending order (lower to higher)
```

```python
# A function that returns the length of the value:
def myFunc(o):
  return len(o)
cars = ['rolls roys', 'ford', 'thar', 'skoda'] # 
cars.sort(key=myFunc)
cars
```
